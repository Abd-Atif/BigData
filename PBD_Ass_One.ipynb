```python
#    Question - 1

def prime(n):
    flag=1
    if (n > 1):
        for i in range(2,n):
            if (n % i) == 0:
                flag=0
                break
            else:
                flag=1
    else:
        return 0
    if flag==1:
        return 1
    else:
        return 0

def PL(L):
    ckPrime=1
    for check in range(len(L)):
            if(prime(L[check])):
                ckPrime=1 
            else:
                ckPrime=0
                break
    return ckPrime

def pL(L):
    pVar=[]
    for i in range(1,L):
        if(prime(i)):
            pVar.append(i)
    return pVar

tmp=0
cL=1
userL = pL(100)
for num in range(0,len(userL)):
    rotatedL =[]
    length = len(str(userL[num]))
    for i in range(0,length):
        newNum = int(str(userL[num])[i:] + str(userL[num])[:i])
        rotatedL.append(newNum)
    cL=PL(rotatedL)
    tmp=tmp+cL
    if (cL==1):
        print(userL[num] ," , ", end = " ")
            
print("\nCount of circular primes:",tmp)
```

```python
#    Question - 5

class MyCircularQueue():
    def __init__(self, k):
        self.k = k
        self.queue = [None] * k
        self.head = self.tail = -1

    def enqueue(self, data):

        if ((self.tail + 1) % self.k == self.head):
            print("The circular queue is full\n")

        elif (self.head == -1):
            self.head = 0
            self.tail = 0
            self.queue[self.tail] = data
        else:
            self.tail = (self.tail + 1) % self.k
            self.queue[self.tail] = data

    def isEmpty(self):
        if self.tail == -1:
            return bool(True)
        return bool(False)
   
    def isFull(self):
        if ((self.tail + 1) % self.k == self.head):
            return bool(True)
        return bool(False)
 
    def dequeue(self):
        if (self.head == -1):
            print("The circular queue is empty\n")

        elif (self.head == self.tail):
            temp = self.queue[self.head]
            self.head = -1
            self.tail = -1
            return temp
        else:
            temp = self.queue[self.head]
            self.head = (self.head + 1) % self.k
            return temp

    def getFront(self):
        if (self.head == -1):
            print("The circular queue is empty\n")
        else:
            return self.queue[self.tail]
        
    def printCQueue(self):
        if(self.head == -1):
            print("No element in the circular queue")

        elif (self.tail >= self.head):
            for i in range(self.head, self.tail + 1):
                print(self.queue[i], end=" ")
            print()
        else:
            for i in range(self.head, self.k):
                print(self.queue[i], end=" ")
            for i in range(0, self.tail + 1):
                print(self.queue[i], end=" ")
            print()

            
size=10
currentElement=0
CQ=MyCircularQueue(size)
mini=999999
CQ.enqueue(00)
CQ.enqueue(11)
CQ.enqueue(60)
CQ.enqueue(33)
CQ.enqueue(99)
CQ.enqueue(88)
CQ.enqueue(22)
CQ.enqueue(34)

CQ.printCQueue()

if not CQ.isEmpty():
    
    start = CQ.dequeue()
    mini = currentElement
    end = CQ.getFront()

    while (start != end):
        
        if start < mini:
            mini = start
        
        CQ.enqueue(start)
        if not CQ.isEmpty():
            start = CQ.dequeue()
    CQ.enqueue(start)

    print("Minimum value:", mini)
    CQ.printCQueue()
else:
    print("Queue is empty")
```

```python
#    Question - 6

class Queue:
    def __init__(self):
        self.items = []
        self.size = 0
 
    def is_empty(self):
        return self.items == []
 
    def enqueue(self, data):
        self.size += 1
        self.items.append(data)
 
    def dequeue(self):
        self.size -= 1
        return self.items.pop(0)
 
    def get_size(self):
        return self.size
    
    def dis(self):
        if self.size>0:
            print(self.items)


class Stack:
    def __init__(self):
        self.Q1 = Queue()
    
    def is_empty(self):
        return self.Q1.is_empty()
    
    def push(self, data):
        self.Q1.enqueue(data)
    
    def pop(self):
        if not self.Q1.is_empty():
            for _ in range(self.Q1.get_size() - 1):
                dequeued = self.Q1.dequeue()
                self.Q1.enqueue(dequeued)
            return self.Q1.dequeue()
        else:
            return "queue is empty"
    
    def displayy(self):
        if not (self.Q1.is_empty()):
            self.Q1.dis()
        else:
            print("stack is empty")

s=Stack()
s.push(13)
s.push(22)
s.push(31)
s.displayy()
print("\npoped:",s.pop())
print("poped:",s.pop())
print("poped:",s.pop(),"\n")
s.displayy()
```
